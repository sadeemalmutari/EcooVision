```python
import random
import pandas as pd
from datetime import datetime, timedelta

people = ["A", "B", "C", "D"]
start_date = datetime(2023, 1, 1)
days = 365
data = []

for day in range(days):
    current_date = start_date + timedelta(days=day)
    day_of_week = current_date.weekday()  # Monday=0, Sunday=6
    
    for person in people:
        # مثال: لو اليوم يوم عمل
        if day_of_week < 5:  # Monday-Friday
            # توليد وقت خروج
            exit_time = current_date + timedelta(
                hours=7 + random.uniform(0, 1),  # من 7 إلى 8 صباحاً
            )
            data.append([person, "Exit", exit_time])
            
            # توليد وقت دخول
            enter_time = current_date + timedelta(
                hours=16 + random.uniform(0, 1),  # من 16 إلى 17 مساءً
            )
            data.append([person, "Enter", enter_time])
        else:
            # عطلة نهاية الأسبوع؛ احتمالية خروجه أقل أو في وقت مختلف
            # لنفرض 50% أن يخرج
            if random.random() < 0.5:
                exit_time = current_date + timedelta(
                    hours=9 + random.uniform(0, 3),  # من 9 إلى 12 ظهراً
                )
                data.append([person, "Exit", exit_time])
                
                enter_time = exit_time + timedelta(
                    hours=3 + random.uniform(0, 2)  # يرجع بعد 3~5 ساعات
                )
                data.append([person, "Enter", enter_time])
            # في حال لم يخرج، لا نضيف سجل

df = pd.DataFrame(data, columns=["PersonID", "Event", "Timestamp"])
df = df.sort_values(by=["Timestamp"]).reset_index(drop=True)
df.head(20)
```
